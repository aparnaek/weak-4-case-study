# Plot of male vs female buyers participated in the sale

```python
data['Buyer Gender'].value_counts()
```

```python
sns.set(font_scale=1.4,)
data['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0,)
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title("Count of Male and Female Buyers", y=1.02);
```

Insight:-In the car sale 5052 were male and 4948 were female buyers.So that males bought more car tha females
